```python
# ===============================
# 1. ORTAM & DRIVE AYARLARI
# ===============================

# Temel kГјtГјphaneler
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt

# TensorFlow / Keras
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

# Colab Drive
from google.colab import drive

# -------------------------------
# Google Drive baДҹla
# -------------------------------
drive.mount('/content/drive')

# -------------------------------
# Proje yollarДұ
# -------------------------------
PROJECT_ROOT = "/content/drive/My Drive/BoneAgeProject"
DATASET_DIR  = os.path.join(PROJECT_ROOT, "DATASET")
MODEL_DIR    = os.path.join(PROJECT_ROOT, "models")

# -------------------------------
# Dataset alt klasГ¶rleri
# -------------------------------
Articular_Surface_Train = os.path.join(DATASET_DIR, "Articular_Surface_Train")
Articular_Surface_Val   = os.path.join(DATASET_DIR, "Articular_Surface_Val")

Epiphysis_Train = os.path.join(DATASET_DIR, "Epiphysis_Train")
Epiphysis_Val   = os.path.join(DATASET_DIR, "Epiphysis_Val")

train_csv = os.path.join(DATASET_DIR, "train.csv")
val_csv   = os.path.join(DATASET_DIR, "val.csv")

# -------------------------------
# Global sabitler
# -------------------------------
IMG_SIZE   = 320
BATCH_SIZE = 8
MAX_AGE    = 240

# -------------------------------
# Ortam bilgisi (kontrol amaГ§lДұ)
# -------------------------------
print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices("GPU"))
print("Proje dizini:", PROJECT_ROOT)
print("Dataset dizini:", DATASET_DIR)
print("Model kayДұt dizini:", MODEL_DIR)
```

```python
# ===============================
# 2. VERД° YГңKLEME & tf.data
# ===============================

def load_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img.astype(np.float32)
    return img


def create_dataset(csv_path, art_dir, epi_dir):
    df = pd.read_csv(csv_path)

    X_art, X_epi, males, y = [], [], [], []

    for _, row in df.iterrows():
        _id = str(row["id"])
        boneage = float(row["boneage"])

        art_path = os.path.join(art_dir, _id + ".png")
        epi_path = os.path.join(epi_dir, _id + ".png")

        if not (os.path.exists(art_path) and os.path.exists(epi_path)):
            continue

        X_art.append(load_image(art_path))
        X_epi.append(load_image(epi_path))
        males.append(1.0 if row["male"] else 0.0)
        y.append(boneage)

    return (
        np.array(X_art, dtype=np.float32),
        np.array(X_epi, dtype=np.float32),
        np.array(males, dtype=np.float32).reshape(-1, 1),
        np.array(y, dtype=np.float32),
    )


# -------------------------------
# Datasetleri yГјkle
# -------------------------------
X_art_train, X_epi_train, male_train, y_train = create_dataset(
    train_csv, Articular_Surface_Train, Epiphysis_Train
)

X_art_val, X_epi_val, male_val, y_val = create_dataset(
    val_csv, Articular_Surface_Val, Epiphysis_Val
)

print("Veriler yГјklendi.")


# -------------------------------
# YAЕһ NORMALД°ZASYONU
# -------------------------------
y_train = y_train / MAX_AGE
y_val   = y_val   / MAX_AGE

# -------------------------------
# tf.data Datasetleri
# -------------------------------
train_ds = tf.data.Dataset.from_tensor_slices(
    (
        {
            "art_input": X_art_train,
            "epi_input": X_epi_train,
            "male_input": male_train,
        },
        y_train,
    )
)

val_ds = tf.data.Dataset.from_tensor_slices(
    (
        {
            "art_input": X_art_val,
            "epi_input": X_epi_val,
            "male_input": male_val,
        },
        y_val,
    )
)

train_ds = train_ds.shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_ds   = val_ds.shuffle(100).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print("tf.data pipeline hazДұr.")
```

```python
# ===============================
# 3. MODEL TANIMI
# ===============================

# --- EfficientNet Backbone ---
def build_backbone(name):
    inp = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = preprocess_input(inp)

    base = EfficientNetB0(
        include_top=False,
        weights="imagenet",
        pooling="avg"
    )

    # 1) TГјm katmanlarДұ dondur
    for layer in base.layers:
        layer.trainable = False

    # 2) Son ~20 katmanДұ (BN hariГ§) aГ§
    for layer in base.layers[-20:]:
        if not isinstance(layer, tf.keras.layers.BatchNormalization):
            layer.trainable = True

    x = base(x)
    return models.Model(inp, x, name=name)


# --- Ana Model ---
def build_model():
    # GiriЕҹler
    art_input   = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name="art_input")
    epi_input   = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name="epi_input")
    male_input  = layers.Input(shape=(1,), name="male_input")

    # Backbone'lar
    branch_art = build_backbone("EffNet_Art")
    branch_epi = build_backbone("EffNet_Epi")

    feat_art = branch_art(art_input)
    feat_epi = branch_epi(epi_input)

    # BirleЕҹtirme
    fused = layers.Concatenate(name="fusion")(
        [feat_art, feat_epi, male_input]
    )

    # Dense bloklarДұ
    x = layers.Dense(512,activation="relu")(fused)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(128,activation="relu",)(x)
    x = layers.Dropout(0.2)(x)

    # ГҮДұkДұЕҹ (0вҖ“1 arasДұ)
    out = layers.Dense(1, activation="sigmoid")(x)

    return models.Model(
        inputs=[art_input, epi_input, male_input],
        outputs=out
    )


# Modeli oluЕҹtur ve Г¶zetle
model = build_model()
model.summary()
```

```python
# ===============================
# 4. COMPILE & EДһД°TД°M
# ===============================

model.compile(
    optimizer=tf.keras.optimizers.AdamW(
        learning_rate=5e-4,
        weight_decay=1e-5
    ),
    loss="mae",
)


# --- Callback'ler ---

#--------Д°yileЕҹmezse durdur
cb_early = EarlyStopping(
    monitor="val_loss",
    patience=20,
    mode="min",
    restore_best_weights=True,
    verbose=1,
)

#-----Г–Дҹrenme hДұzДұnДұ dГјЕҹГјr-----
cb_lr = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.6,
    patience=4,
    mode="min",
    min_lr=1e-6,
    verbose=1,
)

# --- MODELД° EДһД°T ---
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    callbacks=[cb_early, cb_lr],
    verbose=1,
)

#----Modeli Kaydet--------
print("\nEДҹitim tamamlandДұ.")

model_path = os.path.join(
    MODEL_DIR, "bone_age_final_model.keras"
)

model.save(model_path)

print("Final model kaydedildi:")
print(model_path)
```

```python
# =========================================================
# 5. INFERENCE & SUNUM (FINAL MODEL - SADECE TAHMД°N)
# =========================================================

import os
import cv2
import math
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from tensorflow.keras.models import load_model

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

print("рҹ–ҘпёҸ TEST / SUNUM MODU BAЕһLATILDI")

# ------------------ AYARLAR ------------------
IMG_SIZE = 320
BATCH_SIZE = 8
MAX_AGE = 240.0
PER_PAGE = 4

# ------------------ SABД°T DRIVE YOLLARI ------------------
BASE_PROJECT = "/content/drive/My Drive/BoneAgeProject"

MODEL_PATH = os.path.join(
    BASE_PROJECT,
    "models",
    "bone_age_final_model.keras"
)

DATASET_BASE = os.path.join(BASE_PROJECT, "DATASET")

ART_DIR = os.path.join(DATASET_BASE, "Articular_Surface_Test")
EPI_DIR = os.path.join(DATASET_BASE, "Epiphysis_Test")
TEST_CSV = os.path.join(DATASET_BASE, "test.csv")

# ------------------ KONTROLLER ------------------
if not os.path.exists(MODEL_PATH):
    raise RuntimeError(f"Model bulunamadДұ: {MODEL_PATH}")

if not os.path.exists(TEST_CSV):
    raise RuntimeError("test.csv bulunamadДұ!")

print("рҹ§  Model yolu :", MODEL_PATH)
print("рҹ“Ӯ Dataset yolu:", DATASET_BASE)

# ------------------ YARDIMCI (MODELLE UYUMLU) ------------------
def load_img(path):
    img = cv2.imread(path)
    if img is None:
        return None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    return img.astype(np.float32)

# ------------------ VERД° YГңKLE ------------------
df = pd.read_csv(TEST_CSV)

X_art, X_epi, males, y_true, ids = [], [], [], [], []

for _, row in df.iterrows():
    pid = str(row["id"])

    p_art = os.path.join(ART_DIR, pid + ".png")
    p_epi = os.path.join(EPI_DIR, pid + ".png")

    if not (os.path.exists(p_art) and os.path.exists(p_epi)):
        continue

    img_art = load_img(p_art)
    img_epi = load_img(p_epi)

    if img_art is None or img_epi is None:
        continue

    X_art.append(img_art)
    X_epi.append(img_epi)
    males.append(1.0 if row["male"] else 0.0)
    y_true.append(float(row["boneage"]))
    ids.append(pid)

X_art = np.array(X_art)
X_epi = np.array(X_epi)
males = np.array(males).reshape(-1, 1)
y_true = np.array(y_true)

print(f"вң… Toplam test Г¶rneДҹi: {len(y_true)}")

# ------------------ MODEL YГңKLE ------------------
model = load_model(MODEL_PATH, compile=False)
print("рҹҸҶ Final model yГјklendi")

# ------------------ TAHMД°N ------------------
preds = model.predict(
    {
        "art_input": X_art,
        "epi_input": X_epi,
        "male_input": males
    },
    batch_size=BATCH_SIZE,
    verbose=0
).flatten()

pred_ages = preds * MAX_AGE
errors = np.abs(pred_ages - y_true)
ss_res = np.sum((y_true - pred_ages) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
r2 = 1 - (ss_res / ss_tot)
rmse = np.sqrt(np.mean((pred_ages - y_true) ** 2))

# ------------------ ГҮД°ZGД° GRAFД°ДһД° ------------------
plt.figure(figsize=(12, 5))
x = np.arange(1, len(y_true) + 1)

plt.plot(x, y_true, label="GerГ§ek YaЕҹ", linewidth=2)
plt.plot(x, pred_ages, label="Tahmini YaЕҹ", linewidth=2)

plt.xlabel("KiЕҹi Д°ndeksi")
plt.ylabel("YaЕҹ (ay)")
plt.title("GerГ§ek ve Tahmin Edilen Kemik YaЕҹДұ (Test KГјmesi)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ------------------ METД°NSEL Г–ZET ------------------
print("\nрҹ“Ҡ Г–ZET")
print(f"Test MAE : {errors.mean():.2f} ay")
print(f"Test RMSE: {rmse:.2f}")
print(f"Test RВІ  : {r2:.2f}\n")



# ------------------ GГ–RSEL SAYFALAMA (SUNUM STД°LД° - SIKI) ------------------
total = len(ids)
pages = math.ceil(total / PER_PAGE)

for p in range(pages):
    start = p * PER_PAGE
    end = min(start + PER_PAGE, total)

    fig = plt.figure(figsize=(13, 3.4 * (end - start)))

    gs = GridSpec(
        end - start,
        3,
        width_ratios=[1.45, 1.45, 1.45],
        wspace=0.015
    )

    for i, idx in enumerate(range(start, end)):
        # ---------- ART ----------
        ax_art = fig.add_subplot(gs[i, 0])
        ax_art.imshow(X_art[idx] / 255.0)
        ax_art.set_title("Articular Surface", fontsize=12, pad=6)
        ax_art.axis("off")

        # ---------- EPI ----------
        ax_epi = fig.add_subplot(gs[i, 1])
        ax_epi.imshow(X_epi[idx] / 255.0)
        ax_epi.set_title("Epiphysis", fontsize=12, pad=6)
        ax_epi.axis("off")

        # ---------- SONUГҮ ----------
        ax_txt = fig.add_subplot(gs[i, 2])
        ax_txt.axis("off")

        err = errors[idx]

        if err < 7:
            color = "lime"
            status = "(MГңKEMMEL)"
        elif err < 13:
            color = "purple"
            status = "(Д°YД°)"
        else:
            color = "red"
            status = "(SAPMA)"

        ax_txt.text(
            0.5, 0.5,
            f"ID: {ids[idx]}\n\n"
            f"GerГ§ek: {y_true[idx]:.0f} ay\n"
            f"Tahmin: {pred_ages[idx]:.0f} ay\n"
            f"Hata: {errors[idx]:.0f} ay {status}",
            ha="center",
            va="center",
            fontsize=13,
            fontweight="bold",
            color=color,
            bbox=dict(
                facecolor="#ffffff",
                alpha=1,
                pad=10,
                edgecolor="none"
            )
        )



    plt.show()

# ------------------ SONUГҮ TABLOSU ------------------
from IPython.display import display, HTML

results_df = pd.DataFrame({
    "id": ids,
    "gerГ§ek_yaЕҹ(ay)": y_true.astype(int),
    "tahmini_yaЕҹ(ay)": np.round(pred_ages).astype(int),
    "fark(ay)": np.round(errors).astype(int)
})

print(f"рҹ“Ӣ TEST SONUГҮ TABLOSU (Toplam {len(results_df)} kayДұt)")
display(HTML(results_df.to_html(index=False, border=1)))
```
